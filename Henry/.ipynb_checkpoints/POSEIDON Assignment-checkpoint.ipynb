# Environment

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CT_scan_df = pd.read_csv(r'.\datasets\GeoB14403-8_CT_scan.tab', sep='\t', skiprows=16)
direct_shear_df = pd.read_csv(r'.\datasets\GeoB14403-8_drained_direct-shear.tab', sep='\t', skiprows=18)
chem_prop_df = pd.read_csv(r'.\datasets\GeoB14403-8_physical_chemical_properties.tab', sep='\t', skiprows=15)
```

# I - CT Scan Data

```python
CT_scan_df.head()
```

```python
plt.rcParams["figure.figsize"] = (6,8)
plt.plot(CT_scan_df["Radiodensity [HU]"], CT_scan_df["Depth sed [m]"], 
                 color='blue', label="Radiodensity")
plt.ylim(28.38,28.18)
plt.title("Radiodensity")
plt.xlabel("Radiodensity [HU]")
plt.ylabel("Depth [m]")
plt.legend(loc="best")
plt.grid()

plt.show()
```

```python
plt.rcParams["figure.figsize"] = (6,8)
plt.plot(CT_scan_df["Mesoporosity [%]"], CT_scan_df["Depth sed [m]"], 
                 color='red', label="Mesoporosity")

plt.ylim(28.38,28.18)
plt.title("Mesoporosity")
plt.xlabel("Mesoporosity [%]")
plt.ylabel("Depth [m]")
plt.legend(loc="best")
plt.grid()

plt.show()
```

# II - Direct Shear Data

```python
direct_shear_df.head()
```

```python
plt.rcParams["figure.figsize"] = (8,6)
plt.plot(direct_shear_df["DIS [mm] (DS I, horizontal)"], direct_shear_df["Shear str [kPa] (DS I)"], label= "DS I")
plt.plot(direct_shear_df["DIS [mm] (DS II, horizontal)"], direct_shear_df["Shear str [kPa] (DS II)"], label= "DS II")
plt.plot(direct_shear_df["DIS [mm] (DS III, horizontal)"], direct_shear_df["Shear str [kPa] (DS III)"], label= "DS III")


plt.xlim(0,12)
plt.ylim(0,160)

plt.title("Direct Shear")
plt.xlabel("Horizontal displacement [mm]")
plt.ylabel("$\\tau$ [kPa]")
plt.legend(loc="best")
plt.grid()
plt.show()
```

# III - Chem Prop Data

```python
chem_prop_df.head()
```

```python
density_df = chem_prop_df[["Depth sed [m]", "Density [g/cm**3] (bulk density (MSCL), rcorr)"]].dropna()
ca_ti_df = chem_prop_df[["Depth sed [m]", "Ca/Ti (log (Ca/Ti))"]].dropna()
```

```python
plt.rcParams["figure.figsize"] = (6,8)
plt.plot(density_df["Density [g/cm**3] (bulk density (MSCL), rcorr)"], density_df["Depth sed [m]"], 
                 color='red', label="Density")

plt.ylim(29,27.8)
plt.title("Density")
plt.xlabel("Density [g/cm3]")
plt.ylabel("Depth [m]")
plt.legend(loc="best")
plt.grid()

plt.show()
```

```python
plt.rcParams["figure.figsize"] = (6,8)
plt.plot(ca_ti_df["Ca/Ti (log (Ca/Ti))"], ca_ti_df["Depth sed [m]"], 
                 color='red', label="Ca/Ti")

plt.ylim(29,27.8)
plt.title("Ca/Ti")
plt.xlabel("Ca/Ti")
plt.ylabel("Depth [m]")
plt.legend(loc="best")
plt.grid()

plt.show()
```
