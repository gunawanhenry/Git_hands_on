# I - CT Scan Data

```python
CT_scan_df.head()
```

```python
# Create the figure and axes with two y-axes
fig, ax1 = plt.subplots(figsize=(6, 8))
ax2 = ax1.twinx()

# Plot the first dataset (Radiodensity) on the primary y-axis
line1 = ax1.plot(CT_scan_df["Radiodensity [HU]"], CT_scan_df["Depth sed [m]"], 
                 color='blue', label="Radiodensity")

# Plot the second dataset (Mesoporosity) on the secondary y-axis
line2 = ax2.plot(CT_scan_df["Mesoporosity [%]"], CT_scan_df["Depth sed [m]"], 
                 color='red', label="Mesoporosity")

# Set labels and title
ax1.set_xlabel("Radiodensity [HU]")
ax1.set_ylabel("Depth sed [m]")
ax2.set_xlabel("Mesoporosity [%]")
ax2.set_ylabel("No slice [#]")


ax2.set_xlim(920,1080)

# Add secondary x-axis on top
ax3 = ax1.twiny()
ax3.set_xlabel("Mesoporosity [%]")
ax3.set_xlim(0,40)

# Invert the y-axis for both primary and secondary axes
ax1.invert_yaxis()
ax2.invert_yaxis()

# Combine legends from both axes
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc="best")

plt.title("CT Scan Data")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# II - Direct Shear Data

```python
direct_shear_df.head()
```

```python
plt.plot(direct_shear_df["DIS [mm] (DS I, horizontal)"], direct_shear_df["Shear str [kPa] (DS I)"], label= "DS I")
plt.plot(direct_shear_df["DIS [mm] (DS II, horizontal)"], direct_shear_df["Shear str [kPa] (DS II)"], label= "DS II")
plt.plot(direct_shear_df["DIS [mm] (DS III, horizontal)"], direct_shear_df["Shear str [kPa] (DS III)"], label= "DS III")


plt.xlim(0,12)
plt.ylim(0,160)

plt.title("Direct Shear")
plt.xlabel("Horizontal displacement [mm]")
plt.ylabel("$\\tau$ [kPa]")
plt.legend(loc="best")
plt.grid()
plt.show()
```
